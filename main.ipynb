```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data= pd.read_csv('Sub_Division_IMD_2017.csv')
```

```python
data.head()
```

```python
pd.DataFrame(data.SUBDIVISION.value_counts())
```

```python
subdivs=(data.SUBDIVISION.value_counts().index.values)
subdivs.sort()
```

```python
subdivs
```

```python
data_dict={}
for key in subdivs:
    value=pd.DataFrame(columns=data.columns[1:])
    for i in data.index:
        if data.iloc[i,0]== key:
            value=value.append(
                data.iloc[i,1:], ignore_index=True
            )
    data_dict[key]=value
```

```python
data_dict.keys()
```

```python
data_dict['Himachal Pradesh'].head()
```

```python
plt.figure(figsize=(25,15))
for subdiv, i in zip(subdivs,range(1,37)):
    plt.subplot(6,6,i)
    plt.plot(data_dict[subdiv]['YEAR'].values,data_dict[subdiv]['ANNUAL'].values)
    plt.xticks(())
    plt.xlabel(subdiv)
plt.show()
```

```python
div_id={}
for key,val in zip(subdivs,range(len(subdivs))):
    div_id[key]=val
month_id={}
for key,val in zip(data.columns[2:-5], range(len(data.columns[2:-5]))):
    month_id[key]=val
```

```python
div_id, month_id
```

```python
div_id_list=[]
for div in data['SUBDIVISION'].values:
    div_id_list.append(div_id[div])
```

```python
data_with_id=data
data_with_id['SUBDIVISION']=div_id_list
```

```python
data_with_id_for_transformation=data_with_id.drop(columns=['ANNUAL','JF','MAM','JJAS','OND'])
```

```python
data_with_id_for_transformation.head()
```

```python
dnp=np.array([[]])
for i in data_with_id_for_transformation.index:
    subdiv=data_with_id_for_transformation.iloc[i,0]
    year=data_with_id_for_transformation.iloc[i,1]
    for month in range(12):
        arr=np.array([i,year,month,data_with_id_for_transformation.iloc[i,month+2]])
        dnp=np.append(dnp,arr)
```

```python
numpy_data=dnp.reshape(int(dnp.shape[0]/4),4)
```

```python
prepData=pd.DataFrame(numpy_data,columns=['s_div','year','month','rainfall'])
```

```python
prepData.head()
```

```python
prepData.rainfall.isnull().value_counts()
```

```python
prepData.rainfall=prepData.rainfall.fillna(value=0)
```

```python
prepData.rainfall.isnull().value_counts()
```

```python
print("Shapes Comaparison:",data_with_id_for_transformation.shape,'Vs',prepData.shape)
```

```python
plt.figure()
plt.imshow(prepData.corr().values)
plt.title("Pearson's Coorelation Matrix")
plt.xticks((range(4)),prepData.columns)
plt.yticks((range(4)),prepData.columns)
plt.show()
```

```python
X=prepData.iloc[:,:-1].values
y=prepData['rainfall'].values
```

```python
X.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.25)
```

```python
X_train.shape,X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=4)
```

```python
X_train_poly=poly.fit_transform(X_train)
X_test_poly=poly.transform(X_test)
```

```python
lr.fit(X_train_poly,y_train)
lr.score(X_test_poly,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model= RandomForestRegressor(n_estimators=10)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
plt.figure()
plt.bar(prepData.columns[:-1],model.feature_importances_)
plt.show()
```
